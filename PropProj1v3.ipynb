```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
gamma = 1.4
cp = 1.005  # kJ/(kgK)
T1 = 300  # K
T3 = 1800 # K

# Pressure ratio range from 2 to 70
P_ratio = np.linspace(2, 70, 100)

# Calculate thermal efficiency
eta = 1 - 1/((P_ratio) ** ((gamma - 1) / (gamma)))

# Calculate specific power
W_spec = cp * T1 * ((T3/T1) - (T3/T1)*(1/P_ratio) ** ((gamma - 1) / gamma) - (P_ratio) ** ((gamma - 1) / gamma) + 1)

# Create a plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot thermal efficiency on the primary y-axis
ax1.set_xlabel("Pressure Ratio (P2/P1)")
ax1.set_ylabel("Thermal Efficiency", color="blue")
ax1.plot(P_ratio, eta, label="Thermal Efficiency", color="blue", linewidth=2)
ax1.tick_params(axis="y", labelcolor="blue")

# Create a second y-axis for specific power
ax2 = ax1.twinx()
ax2.set_ylabel("Specific Power (kJ/kg)", color="red")
ax2.plot(P_ratio, W_spec, label="Specific Power", color="red", linewidth=2)
ax2.tick_params(axis="y", labelcolor="red")

# Title and grid
plt.title("Thermal Efficiency & Specific Power vs Pressure Ratio")
fig.tight_layout()
plt.grid(True)
plt.show()
```
